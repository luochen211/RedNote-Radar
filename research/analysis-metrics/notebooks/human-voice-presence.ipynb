```python
#导入相关库
from faster_whisper import WhisperModel
import os
import pandas as pd
from tqdm import tqdm
```

```python
#导入模型
model_size = "large-v3"
model = WhisperModel(model_size, device="cuda", compute_type="float16")
```

```python
#数据文件夹路径
folder_path = "./data/65814f7a0000000016004ad1.mp3"
#存储识别结果
results = []
#遍历文件中的文件
for filename in tqdm(os.listdir(folder_path)):
    if filename.endswith(".mp3"):  #识别mp3类型文件
        audio_url = os.path.join(folder_path, filename) #文件路径

        segments, info = model.transcribe(audio_url, beam_size=5)  #识别文字
        #print("Detected language '%s' with probability %f" % (info.language, info.language_probability))
        segments_list = []
        for segment in segments:
            if segment.no_speech_prob>0.6  #准确率判断
            segment_dict = {
                'start': segment.start,
                'end': segment.end,
                'text': segment.text
            }
            segments_list.append(segment_dict)
        results.append({   #识别结果存储
            "file_name": filename,
            'language': info.language,
            'probability':info.language_probability,
            'text_info':segments_list
        })
print('Audio extraction completed.')
df = pd.DataFrame(results, columns=["file_name","language" ,"probability", "text_info"])
```
